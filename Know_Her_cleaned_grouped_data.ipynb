```python
data_set.rename(columns={'Demographics Question':'Demographics_Question', 'Demographics Response': 'Demographics_Response', 'Survey Year':'Survey_Year'}, inplace=True)
print(data_set.columns)
```

Step2: Sort valuable data out. This step, I am going to select three columns to analyze, survey year, country, and value, and sort data according to survey year in ascending order. 

```python
data_sort = data_set[['Survey_Year', 'Country', 'Gender', 'Value']]
data_sort.sort_values('Survey_Year').head(60)
```

Step3: Group data by country. This step, I want to see which country has over one year data to show trends, but I find out only one country has one year data. There are total 70 countries.

```python
country_grp = data_sort.groupby(['Country'])
country_grp['Survey_Year'].value_counts()
```

Step4: Group data by year. This step, I want to see which year has the most countries' data.

```python
year_grp = data_sort.groupby('Survey_Year')
year_grp_counts = year_grp['Country'].value_counts()
year_grp_counts
```

Step5: Group by year and country, calculate median values. Group two columns and calculate median value of column 'Value', and round decimals into 2.

```python
year_country_grp = data_sort.groupby(['Survey_Year', 'Country', 'Gender'], as_index=False)
cleaned_data = year_country_grp.mean('Value').round(2)
cleaned_data
```

```python
cleaned_data.rename(columns={'Value': 'Average_Percentage_of_GBV_Acceptance'}, inplace=True)
cleaned_data
```

```python
cleaned_data.to_csv('data_2000_2018_Gender_Based_Violence_Acceptance.csv', index = False, header = True)
```
